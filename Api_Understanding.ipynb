```python
import requests
from datetime import datetime, timedelta

SYMBOL = "TSLA"  # or "SPY" for S&P 500 ETF
START_DATE = "2023-01-01"
END_DATE = "2024-01-01"

# --------- Function to fetch EOD prices ----------
def fetch_eod(symbol, start_date, end_date):
    url = "https://api.stockdata.org/v1/data/eod"
    params = {
        "symbols": symbol,
        "api_token": API_KEY,
        "date_from": start_date,
        "date_to": end_date
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json().get("data", [])
    else:
        print("EOD Error:", response.status_code, response.text)
        return []

# --------- Function to fetch news ----------
def fetch_news(symbol, start_date, end_date):
    url = "https://api.stockdata.org/v1/news/all"
    params = {
        "symbols": symbol,
        "api_token": API_KEY,
        "date_from": start_date,
        "date_to": end_date,
        "language": "en"
    }
    response = requests.get(url, params=params)
    if response.status_code == 200:
        return response.json().get("data", [])
    else:
        print("News Error:", response.status_code, response.text)
        return []

# --------- Batch retrieval (optional for very long ranges) ----------
def daterange(start, end, step_days=30):
    start_dt = datetime.strptime(start, "%Y-%m-%d")
    end_dt = datetime.strptime(end, "%Y-%m-%d")
    current = start_dt
    while current < end_dt:
        yield current, min(current + timedelta(days=step_days-1), end_dt)
        current += timedelta(days=step_days)

# --------- Collect EOD and news in batches ----------
all_prices = []
all_news = []

for start, end in daterange(START_DATE, END_DATE):
    start_str = start.strftime("%Y-%m-%d")
    end_str = end.strftime("%Y-%m-%d")
    all_prices += fetch_eod(SYMBOL, start_str, end_str)
    all_news += fetch_news(SYMBOL, start_str, end_str)

# --------- Print sample ----------
print("Sample EOD Prices:")
for p in all_prices[:5]:
    print(p)

print("\nSample News:")
for n in all_news[:5]:
    print(f"{n['published_at']} - {n['title']}")
```

```python
import requests
import json
import pandas as pd

API_KEY = "855543b3-48e5-4b6c-825b-66ef1cdfe23d"  # replace with your NewsAPI.ai key

def get_company_news(company="Apple Inc", date_start="2025-01-01", date_end="2025-09-11"):
    url = "https://newsapi.ai/api/v1/article/getArticles"

    payload = {
        "action": "getArticles",
        "query": {
            "$query": {
                "$and": [
                    {
                        "dateStart": date_start,
                        "dateEnd": date_end
                    },
                    {
                        "$or": [
                            {"keyword": company}
                        ]
                    },
                    {
                        "$or": [
                            {"sourceUri": "wsj.com"},
                            {"sourceUri": "ft.com"},
                            {"sourceUri": "marketwatch.com"},
                            {"sourceUri": "bloomberg.com"},
                            {"sourceUri": "reuters.com"},
                            {"sourceUri": "cnbc.com"}
                        ]
                    }
                ]
            }
        },
        "articlesPage": 1,
        "articlesCount": 100,   # max per request
        "articlesSortBy": "date",
        "articlesSortByAsc": False,
        "articlesArticleBodyLen": -1,
        "resultType": "articles",
        "apiKey": API_KEY
    }

    headers = {"Content-Type": "application/json"}
    response = requests.post(url, headers=headers, data=json.dumps(payload))

    if response.status_code == 200:
        data = response.json()
        articles = data.get("articles", {}).get("results", [])

        # Build table rows
        rows = []
        for article in articles:
            headline = article.get("title", "")
            date = article.get("date", "")
            source = article.get("source", {}).get("title", "")
            rows.append([headline, date, source])

        # Create DataFrame
        df = pd.DataFrame(rows, columns=["Headline", "Date", "Source"])
        return df
    else:
        print("Error:", response.status_code, response.text)
        return pd.DataFrame(columns=["Headline", "Date", "Source"])

# Run the function
df = get_company_news()
print(df)
print(f"\nTotal articles: {len(df)}")
```
